```python
# https://dev.socrata.com/
query = '''
    SELECT * 
    WHERE agency="NYPD" 
        AND created_date > "2024-03-01T00:00:00.000" 
    '''

url = f'https://data.cityofnewyork.us/resource/erm2-nwe9.json?$query={query}'

response = requests.get(url)
data = response.json()

pd.set_option('display.max_columns', None)
data = pd.DataFrame(data)
data
```

```python
# https://dev.socrata.com/
query = '''
    SELECT 
        incident_zip,
        COUNT(unique_key) AS count
    WHERE complaint_type="Illegal Parking"
        AND created_date > "2024-03-01T00:00:00.000" 
        AND borough="BROOKLYN"
    GROUP BY incident_zip
    ORDER BY count DESC
    '''

url = f'https://data.cityofnewyork.us/resource/erm2-nwe9.json?$query={query}'

response = requests.get(url)
data = response.json()

data = pd.DataFrame(data)
data
```
