```python
path = r'C:\Users\cansu\OneDrive\Desktop\web-images'
import os

files = os.walk(path)

# create all full paths
image_paths = []
for root, dirs, files in os.walk(path):
    for file in files:
        # if jpg or png or jpeg
        if file.endswith('.jpg') or file.endswith('.png') or file.endswith('.jpeg'):
            image_paths.append(os.path.join(root, file))
    
image_paths
```

```python
# open each image and check its width and close image
from PIL import Image
new_width = 1920

for image_path in image_paths:
    print( image_path)
    
    with Image.open(image_path) as img:
        width, height = img.size
        ratio = width / height
        
        print( width, height, ratio )

        if width > new_width:
            print( image_path)
            new_height = int(new_width / ratio)
            img = img.resize((new_width, new_height))
            img.save( image_path )
            print( 'Resized' )

            # make rgba
            img = img.convert('RGB')

            # convert to jpg
            image_path = image_path.replace('.png', '.jpg')
            img.save(image_path)
    
```
